 Crear un análisis simple de las ventas de un pequeño negocio, incluyendo la
creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados
con Matplotlib.
Los estudiantes trabajarán con los siguientes datos de ventas mensuales:
Datos de ventas mensuales (en miles)
data = {“Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto",
"Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60,
70, 80, 75, 90],}


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas mensuales (en miles)
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

# 1. Crear un DataFrame
df_ventas = pd.DataFrame(data)
print("DataFrame de Ventas Mensuales:")
print(df_ventas)
print("\n")

# 2. Uso de operaciones de NumPy
# Convertir la columna 'Ventas' a un array de NumPy para operaciones numéricas
ventas_np = np.array(df_ventas['Ventas'])

print("Operaciones de NumPy en las Ventas:")
print(f"Ventas totales (en miles): {np.sum(ventas_np)}")
print(f"Venta promedio (en miles): {np.mean(ventas_np):.2f}")
print(f"Venta máxima (en miles): {np.max(ventas_np)}")
print(f"Venta mínima (en miles): {np.min(ventas_np)}")
print(f"Desviación estándar de ventas (en miles): {np.std(ventas_np):.2f}")
print("\n")

# 3. Visualización de los resultados con Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(df_ventas['Mes'], df_ventas['Ventas'], color='skyblue')
plt.xlabel('Mes')
plt.ylabel('Ventas (en miles)')
plt.title('Ventas Mensuales del Negocio')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Realizar un análisis simple de las calificaciones de estudiantes en diferentes
asignaturas, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la
visualización de los resultados con Matplotlib.
Los estudiantes trabajarán con los siguientes datos de calificaciones:
Datos de calificaciones de estudiantes
data = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah",
"Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80,
88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes
data = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# 1. Crear un DataFrame
df_calificaciones = pd.DataFrame(data)
print("DataFrame de Calificaciones de Estudiantes:")
print(df_calificaciones)
print("\n")

# 2. Uso de operaciones de NumPy
print("Operaciones de NumPy en las Calificaciones por Asignatura:")
for subject in ['Matemáticas', 'Ciencias', 'Historia']:
    notas_np = np.array(df_calificaciones[subject])
    print(f"--- {subject} ---")
    print(f"Calificación promedio: {np.mean(notas_np):.2f}")
    print(f"Calificación máxima: {np.max(notas_np)}")
    print(f"Calificación mínima: {np.min(notas_np)}")
    print(f"Desviación estándar: {np.std(notas_np):.2f}")
    print("\n")

print("Operaciones de NumPy en las Calificaciones por Estudiante (Promedio):")
df_calificaciones['Promedio General'] = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean(axis=1)
print(df_calificaciones[['Estudiante', 'Promedio General']])
print(f"Promedio general de todos los estudiantes: {df_calificaciones['Promedio General'].mean():.2f}")
print("\n")

# 3. Visualización de los resultados con Matplotlib

# Visualización 1: Promedio de calificaciones por asignatura
plt.figure(figsize=(10, 6))
avg_subject_grades = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean().sort_values(ascending=False)
plt.bar(avg_subject_grades.index, avg_subject_grades.values, color=['lightcoral', 'lightgreen', 'lightblue'])
plt.xlabel('Asignatura')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Asignatura')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualización 2: Promedio de calificaciones por estudiante
plt.figure(figsize=(12, 6))
plt.bar(df_calificaciones['Estudiante'], df_calificaciones['Promedio General'], color='lightgray')
plt.xlabel('Estudiante')
plt.ylabel('Calificación Promedio General')
plt.title('Calificación Promedio General por Estudiante')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
